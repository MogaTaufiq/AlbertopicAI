```python
df = pd.read_csv('/content/cleaned_data_for_topic_modeling.csv')

df.head()
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

# Download NLTK resources
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt_tab') # Download the missing 'punkt_tab' resource

# Membuat objek Lemmatizer
lemmatizer = WordNetLemmatizer()

# Fungsi untuk membersihkan teks dengan NLTK
def clean_text_nltk(text):
    # Tokenisasi
    tokens = word_tokenize(text.lower())

    # Pembuangan stopwords
    stop_words = set(stopwords.words('english'))
    filtered_tokens = [word for word in tokens if word not in stop_words and word.isalpha()]

    # Lematisasi
    lemmatized_tokens = [lemmatizer.lemmatize(word) for word in filtered_tokens]

    # Gabungkan kembali menjadi string
    return ' '.join(lemmatized_tokens)

# Terapkan pembersihan pada kolom 'title' dan 'abstract'
df['cleaned_title'] = df['title'].apply(clean_text_nltk)
df['cleaned_abstract'] = df['abstract'].apply(clean_text_nltk)

# Tampilkan beberapa baris pertama setelah pembersihan
df[['title', 'cleaned_title', 'abstract', 'cleaned_abstract']].head()
```
