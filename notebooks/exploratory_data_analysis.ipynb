```python
import pandas as pd
import os
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Manually specify the absolute path to the file
processed_data_path = "/Users/MogaTaufiq/iCloud Drive (Archive)/Documents/Documents - Moga's Book/College/SEM 6/MLOps/AlbertAI/data/processed_data/processed_titles.csv"

# Verify the file path
print("File path:", processed_data_path)
print("File exists:", os.path.exists(processed_data_path))

# Load the preprocessed data
df = pd.read_csv(processed_data_path)

# Join all processed titles into one string
all_text = ' '.join(df['Processed_Text'].dropna())

# Generate the WordCloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(all_text)

# Display the WordCloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis('off')  # Hide axes
plt.show()
```

```python
# Count word frequencies using Pandas' value_counts on tokenized words
from collections import Counter

# Tokenize the words in 'Processed_Text'
all_words = ' '.join(df['Processed_Text']).split()
word_freq = Counter(all_words)

# Get the most common words
common_words = word_freq.most_common(20)

# Create a bar chart for the top 20 words
words, counts = zip(*common_words)
plt.figure(figsize=(12, 6))
plt.bar(words, counts)
plt.xticks(rotation=45, ha="right")
plt.title("Top 20 Most Common Words")
plt.xlabel("Words")
plt.ylabel("Frequency")
plt.show()
```
